## 3.4 Check for Missing Values


One of the most common problems in data science is missing values. To detect them, we can us the method .isnull(). With this method, we can get a boolean series (True or False). 

After we identify missing values we can 

1) leave them as it is

2) drop them with dropna() function

3) fill missing value with fillna() 

4) fill missing values with test statistics like mean

```python
matches.isnull().sum()

#here we are finding the sum of the missing values from every colummn of the World Cup Matches data
```

#### As we can see there are a lot of missing values in the WorldCupMatches.csv data. For this analysis we wil choose to ignore these missing values.

```python
world_cups.isnull().sum()
```

#### As we can see there are no missing data from the WorldCups.csv 

```python
players.isnull().sum()
```

#### In this data frame we can see there a lot of missing values for postion and events. So like which postion the players played, etc. 

#### Further we can find the exact percentage of the 33641 missing 'positions' from the entire data

```python
def missing(players):
    print (round((players.isnull().sum() * 100/ len(players)),2))

missing(players)
```

#### We can see 89 % of data in positions is missing while 76% in positions is missing. For this analysis we shall forget about the missing "Events" as it is not of importance to me. Although I do care about the positions. As mentioned before we can fill in these missing values

```python
# Lets see what "postions" the data has that is not missing

players["Position"].value_counts()
```

#### We can see the data has positions such a GK (Goalkeeper), C (which should be center-forward), and GKC (which I dont know what that position is). For the sake of simplicity, instead of re-naming all the positions, lets just deal with the 33641 missing position. So lets replace the missing positions with "FP" also known as Field player.


```python
players["Position"].fillna("Field Player", inplace = True)
```

* Match outcomes by home team wins 

```python
world_cups ['Attendance'] = world_cups ['Attendance'].str.replace(".", "")
```

```python
def get_labels(matches):
    if matches['Home Team Goals'] > matches['Away Team Goals']:
        return 'Home Team Win'
    if matches['Home Team Goals'] < matches['Away Team Goals']:
        return 'Away Team Win'
    return 'DRAW'
```

```python
matches['outcomes'] = matches.apply(lambda x: get_labels(x), axis = 1)
```

```python
matches.head()
```

```python
mt = matches['outcomes'].value_counts()
mt
```

```python
plt.figure(figsize = (6, 6))

mt.plot.pie(autopct = "%1.0f%%", colors = sns.color_palette('winter_r'), shadow = True)

c = plt.Circle((0, 0), 0.4, color = 'white')
plt.gca().add_artist(c)
plt.title('Match Outcomes by Home and Away Teams')
plt.show()
```
